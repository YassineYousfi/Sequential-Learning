```python
star, V, X = optimal_value_iteration(mdp, 1, 100)
```

### $\epsilon$ greedy Q Learning

```python
V , policy, episode_rewards, N = q_learning(mdp, 300, 100, epsilon=0.01)
```

```python
V , policy, episode_rewards_ucb, N = q_learning_ucb(mdp, 300, 100)
```
